# Function to compare two list  : list_compa(A,B)

```python
def list_compa(A,B):
    k=0
    c=0
    d=0
    num_same = []
    for i in A:    
        for j in B:
            k += 1
            if i == j:
              c += 1
              num_same.append(i)
            else:
              d += 1 
    
    print(f' number of sames ={c}, and it is {num_same}')
    
    
        
    if (c == 0) :
        print('There is nothing A=B')
    elif (c == 1) :
        print('There is 1 equal number in A and B')    
    elif (c == 2) :
        print('There is 2 equal number in A and B')    
    elif (c == 3) :
        print('There is 3 equal number in A and B')
    elif (c == 4) :
        print('There is 4 equal number in A and B')    
    elif (c == 5) :
        print('There is 5 equal number in A and B')
    elif (c == 6) :
        print('The two list is the same, A = B')  
    return num_same
```

# Function to classfy fruits : fruit_pick(f_list)

```python
def fruit_pick(f_list):
    f_num = {}
    for i in f_list:
        count=0
        for j in f_list:
            if (i==j):
                count += 1
            continue
        f_num[i]= count  
    f_sort=sorted(f_num.items(),key=lambda item:item[1], reverse=True) 
    return f_sort;
    print(f_num)
    print(f_sort)
```
